### Day 2 Puzzle 1
Calculate the horizontal position and depth you would have after following the planned course. What do you get if you multiply your final horizontal position by your final depth?

```python
sub_df = pd.read_csv('Data/day2_puzzle1_sub_coordinates.csv', sep = ' ', names = ['Position', 'Measurement']) 
sub_df
```

```python
totals = pd.DataFrame(sub_df['Measurement'].groupby(sub_df['Position']).sum())
totals = totals.reset_index()
totals.columns = ['Position', 'Sum']
totals
```

```python
down = totals['Sum'][totals['Position'] == 'down'][0]
forward = totals['Sum'][totals['Position'] == 'forward'][1]
up = totals['Sum'][totals['Position'] == 'up'][2]
forward_final = forward
vertical_final = down - up

print(forward_final)
print(vertical_final)
```

```python
print(f'\
Final horizontal position: {forward_final}\n\
Final vertical position: {vertical_final}\n\
Final Position Vector: {forward_final*vertical_final:,}')
```

Solution

```python
#Using narimiran solution
data = map(str.split, open('Data/day2_puzzle2.txt').readlines())
horizontal = vertical = aim = 0

for command, value in data: 
    value = int(value)
    
    if command == 'forward':
        horizontal = value + horizontal
        vertical = vertical + aim * value
    
    elif command == 'down':
        aim = aim + value
    
    elif command == 'up':
        aim = aim - value

print(f'Aim: {aim}')
print(f'Horizontal: {horizontal}')
print(f'Vertical: {vertical:,}') 
print(f'Horizontal*Aim: {horizontal*aim:,}')
print(f'Horizontal*Vertical: {horizontal*vertical:,}')
```
